### Подготовка csv файлов

Формирование попарно списков рецептов салатов, в которых повторяются ингредиенты из других салатов

```python
a = []
b = []

for i, l in enumerate(df2['ingredients']):
    for el in l.split(','):     
        
        for j in range(len(df2['ingredients'])):
            if el in list(df2['ingredients'])[j].split(',') and i!=j:
                a.append(df2['title'][i])
                b.append(df2['title'][j])
            else:
                pass

            
print(a) 
#print(b)
```

```python
list_tuples = list(zip(a, b)) 
```

```python
df3 = pd.DataFrame(list_tuples, columns=['source', 'target']) 
df3
```

```python
df3['weight'] = [1 for i in range(len(df3))] # добавление нового признака (кол-во повторяемых ингредиентов)
df3['weight']                                 
```

```python
df3
```

```python
df3.duplicated().sum()
```

```python
df4 = pd.DataFrame({'weight' : df3.groupby( ['source', 'target'] ).size()}).reset_index()
df4
# рассчитываем общее число одинаковых ингредиентов для каждой пары рецептов
```

```python
# сохраняем данные по калорийности салатов в файл .csv как узлы графов и  по повторяющимся ингредиентам как ребра
df1.to_csv('nods.csv', index=True)
df4.to_csv('edges.csv', index =False)
```
