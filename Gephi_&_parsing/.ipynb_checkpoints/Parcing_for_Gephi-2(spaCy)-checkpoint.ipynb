```python
import pandas as pd
import numpy as np

import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from pymystem3 import Mystem
#from tqdm import tqdm_notebook
from tqdm.notebook import tqdm
```

```python
# df['ingredients'] = df['ingredients'].str.replace('[^\w\s]', '')
#df['ingredients'] = re.sub(r'[^\w\s]+|[\d]+', r'', df['ingredients'])

df['ingredients'] = df['ingredients'].apply(lambda x: re.sub(r'[^\w\s\,]+|[\d]+ ', '', x))
#df['ingredients'] = df['ingredients'].str.replace('[½ pound]', '')

df['ingredients']
```

```python
lem = Mystem()
nltk.download('stopwords')
stop = stopwords.words('english')

def preprocess_text(x):
    return [word.rstrip() for word in (''.join(lem.lemmatize(x))).split(' ') if (word not in stop and word != '')]
```

```python
stop.extend([
u'¾', u'½', u'⅛', u'⅔', u'cups', u'cup', u'pound', u'tablespoons', u'teaspoons', u'ounce',\
      u'2u200', u'1u200', u'¼', u'heads', u'white', u'red',\
      u'1u2009½', u'black', u'2u2009½', u'1u2009'])

```

```python
text_data = df['ingredients'].to_numpy()

text_data_lemmatized = []
for text in tqdm(text_data):   # notebook.tqdm вместо tqdm_notebook
    text_data_lemmatized.append(preprocess_text(text))
```

```python
for i,k in enumerate(text_data_lemmatized):
    m = ','.join(k)
                         
    text_data_lemmatized[i] = m
    
text_data_lemmatized[1]                         
                         
```

```python
df['ingredients'] = text_data_lemmatized

df['ingredients'][6]
```
