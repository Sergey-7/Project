```python
# df['ingredients'] = df['ingredients'].str.replace('[^\w\s]', '')
#df['ingredients'] = re.sub(r'[^\w\s]+|[\d]+', r'', df['ingredients'])

df['ingredients'] = df['ingredients'].apply(lambda x: re.sub(r'[^\w\s\,]+|[\d]+ ', '', x))
#df['ingredients'] = df['ingredients'].str.replace('[½ pound]', '')

df['ingredients']
```

```python
df['ingredients'] = df['ingredients'].str.replace('¾', '').str.replace('½', '').str.replace('cup', '').str.replace('pound','')\
                             .str.replace('tablespoons', '').str.replace('teaspoons', '').str.replace('ounce', '')\
                              .str.replace('2u2009', '').str.replace('1u2009', '').str.replace(' ¼ ', '')
df['ingredients']
```
