Парсинг данных по рецептам салатов с сайта  'https://www.allrecipes.com/recipes/96/salad/' и запись нужных параметров в файл csv

```python
import json
import logging
from pprint import pprint
from time import sleep

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import re
from elasticsearch import Elasticsearch

```

```python
def search(es_object, index_name, search):
    res = es_object.search(index=index_name, body=search, request_timeout=60)  # по умолч. req_timeout = 10
    pprint(res)

def create_index(es_object, index_name):
    created = False
    # index settings
    settings = {
        "settings": {
            "number_of_shards": 1,
            "number_of_replicas": 0
        },
        "mappings": {
            "salads": {
                "dynamic": "strict",
                "properties": {
                    "title": {
                        "type": "text"
                    },
                    "submitter": {
                        "type": "text"
                    },
                    "description": {
                        "type": "text"
                    },
                    "calories": {
                        "type": "integer"
                    },
                    "ingredients": {
                        "type": "nested",
                        "properties": {"type": "text"}

                        },
                    }
                }
            }
        }

    try:
        if not es_object.indices.exists(index_name):
            # Ignore 400 means to ignore "Index Already Exist" error.
            es_object.indices.create(index=index_name, ignore=400, body=settings)
            #print('Created Index')
        created = True
    except Exception as ex:
        print(str(ex))
    finally:
        return created


def store_record(elastic_object, index_name, record):
    is_stored = True
    try:
        outcome = elastic_object.index(index=index_name, doc_type='salads', body=record)
        #print(outcome)
    except Exception as ex:
        #print('Error in indexing data')
        #print(str(ex))
        is_stored = False
    finally:
        return is_stored


def connect_elasticsearch():
    _es = None
    _es = Elasticsearch([{'host': 'localhost', 'port': 9200}])
    if _es.ping():
        print('Yay Connected')
    else:
        print('Awww it could not connect!')
    return _es


def parse(url):
    title = '-'
    submit_by = '-'
    description = '-'
    #calories = 0
    ingredients = []
    nutrition =[]
    rec = {}
    #print(lst)
    #for i in lst:
    r = requests.get(url, headers=headers)
    #print(r.text)
    #print(r.status_code)
    #if r.status_code == 200:

    u = r.text
    #print(u)
    soup = BeautifulSoup(u, 'html.parser')  # 'lxml')
    #print(soup)
    # title
    title_section = soup.select('.recipe-title')
    #print(title_section)
    # submitter

    submitter_section =  soup.find_all('a', {'class':['author-name link']})
    #print(submitter_section)
    # description

    description_section =  soup.find_all('div', {'class':['recipe-summary margin-8-bottom']})
    #print(description_section)
    ingredient_section = soup.find_all('li', {'class':['ingredients-item']})
    #print(ingredient_section)
    nutrition_section = soup.find_all('div', {'class':['partial recipe-nutrition-section']})
    #print(nutrition_section)
    if title_section:
        title = title_section[0].text # берем элемент <class 'bs4.element.Tag'> списка title_section и выбираем из него текст
        #print(title)
    if description_section:
        description = description_section[0].text.strip().replace('"', '')
        #print(description)
    if submitter_section:
        submit_by = submitter_section[0].text.strip()
        #print(submit_by)
    if ingredient_section:
        for el in ingredient_section:
            ingredient = el.text.strip().replace('"', '')
        #ingredients = ingredient_section[0].text.strip().replace('"', '')
            ingredients.append(ingredient)
        #print(ingredients)
    if nutrition_section:
        nutrition = nutrition_section[0].text.strip()[42:].replace('Full Nutrition','')
        calories = int(nutrition[10: 12])    #int(nutrition[:5])
        #print(calories)
    rec = {'title': title, 'submitter': submit_by, 'description': description, 'ingredients': ingredients, 'calories': calories, 'nutrition': nutrition}

    #print(rec)
    #return json.dumps(rec)


def search(es_object, index_name, search):
    res = es_object.search(index=index_name, body=search, request_timeout=60)  # по умолч. req_timeout = 10
    pprint(res)

def create_index(es_object, index_name):
    created = False
    # index settings
    settings = {
        "settings": {
            "number_of_shards": 1,
            "number_of_replicas": 0
        },
        "mappings": {
            "salads": {
                "dynamic": "strict",
                "properties": {
                    "title": {
                        "type": "text"
                    },
                    "submitter": {
                        "type": "text"
                    },
                    "description": {
                        "type": "text"
                    },
                    "calories": {
                        "type": "integer"
                    },
                    "ingredients": {
                        "type": "nested",
                        "properties": {"type": "text"}

                        },
                    }
                }
            }
        }

    try:
        if not es_object.indices.exists(index_name):
            # Ignore 400 means to ignore "Index Already Exist" error.
            es_object.indices.create(index=index_name, ignore=400, body=settings)
            #print('Created Index')
        created = True
    except Exception as ex:
        print(str(ex))
    finally:
        return created


def store_record(elastic_object, index_name, record):
    is_stored = True
    try:
        outcome = elastic_object.index(index=index_name, doc_type='salads', body=record)
        #print(outcome)
    except Exception as ex:
        #print('Error in indexing data')
        #print(str(ex))
        is_stored = False
    finally:
        return is_stored


def connect_elasticsearch():
    _es = None
    _es = Elasticsearch([{'host': 'localhost', 'port': 9200}])
    if _es.ping():
        print('Yay Connected')
    else:
        print('Awww it could not connect!')
    return _es

lines = []
def parse(url):
    title = '-'
    submit_by = '-'
    description = '-'
    #calories = 0
    ingredients = []
    nutrition =[]
    rec = {}
    #print(lst)
    #for i in lst:
    r = requests.get(url, headers=headers)
    #print(r.text)
    #print(r.status_code)
    #if r.status_code == 200:

    u = r.text
    #print(u)
    soup = BeautifulSoup(u, 'html.parser')  # 'lxml')
    #print(soup)
    # title
    title_section = soup.select('.recipe-title')
    #print(title_section)
    # submitter

    submitter_section =  soup.find_all('a', {'class':['author-name link']})
    #print(submitter_section)
    # description

    description_section =  soup.find_all('div', {'class':['recipe-summary margin-8-bottom']})
    #print(description_section)
    ingredient_section = soup.find_all('li', {'class':['ingredients-item']})
    #print(ingredient_section)
    nutrition_section = soup.find_all('div', {'class':['partial recipe-nutrition-section']})
    #print(nutrition_section)
    if title_section:
        title = title_section[0].text # берем элемент <class 'bs4.element.Tag'> списка title_section и выбираем из него текст
        #print(title)
    if description_section:
        description = description_section[0].text.strip().replace('"', '')
        #print(description)
    if submitter_section:
        submit_by = submitter_section[0].text.strip()
        #print(submit_by)
    if ingredient_section:
        for el in ingredient_section:
            ingredient = el.text.strip().replace('"', '')
        #ingredients = ingredient_section[0].text.strip().replace('"', '')
            ingredients.append(ingredient)
        #print(ingredients)
    if nutrition_section:
        nutrition = nutrition_section[0].text.strip()[15:].replace('Full Nutrition','')
        #print(nutrition)
        calories = nutrition[:2]    #int(nutrition[:5])
        #print(calories)
    rec = {'title': title, 'description': description, 'ingredients': ingredients, 'calories': calories, 'nutrition': nutrition}
    lines.append(rec)
    print(lines)
    import csv
    with open('recipies.csv', 'w', newline='', encoding='utf-8') as csvfile:
        columns = ['title','description', 'ingredients', 'calories', 'nutrition']

        writer = csv.DictWriter(csvfile, fieldnames=columns)

        writer.writeheader()

        for line in lines:
            #writer.writerow((line['title'], line['description'], line['calories'], line['nutrition']))
            writer.writerow(line)


        #return json.dumps(rec)


```
